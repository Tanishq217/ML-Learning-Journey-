# `DATA MANIPLATION USING PANDAS`

```python
# How to get csv file data in panda database !

import pandas as pd

df = pd.read_csv("train.csv")
df
```

```python
df.head()
```

```python
print(r"Tanishq\nSingh")
```

```python
# Filtering the data !

# 1) Index a column

df['Sex'].head()
```

```python
a = 12
a == 12
```

```python
# using this boolean series to filter complete table

df [ df ['Sex'] == 'male'] # we will get only those rows where the sex is male
```

```python
# lets take out all the data where sex is female
```

```python
df[df['Sex'] == 'female']
```

```python
# Applying Multiple Filters

# ONE FILTER >> df [ condition ]
# MANY FILTER >> df [(condition_1) & (condition_2) & (condition_3)]
# AND -> &
# OR -> |
```

```python
# DF ON MALE where sex is m

df[(df['Sex'] == 'male') & (df['Age'] < 25)]
# TWO CONDITIONS > CONDITION 1 and Condition 2
# round brackets are very compulsory !
```

```python
# WE NEED ALL FEMALE WITH FARE GREATER THAN 50$ DOLLERS !

df[(df['Sex'] == 'female') & (df['Fare'] > 50)]
```

```python
# we also need count
df[(df['Sex'] == 'female') & (df['Fare'] > 50)].shape
# this will give us the number of rows and columns
```

```python
# when we need count

df[(df['Sex'] == 'female') & (df['Fare'] > 50)]['PassengerId'].count()
#selects the column PassengerId from the filtered rows.
```

```python
# How to add or delete a column

# ADDING A COLUMN "Double Price" == Fare * 2

df['Double Price'] = df['Fare'] * 2

df.head()
```

```python
# how many category in Embarkes !

df['Embarked'].value_counts()
```

```python
df['Embarked'].isna().sum()
# we have 2  missings
```

```python
# There are 3 categories !


# Lets write a function now.

def em(x):  # takes x as an Input
    if x == 'S':
      return 1
    elif x == 'C' :
      return 2
    elif x == 'Q':
      return 3
    else :
      return 3


# apply this function to column Embarked !

df['Embarked'].apply(em)
```

```python
df['em-c'] = df['Embarked'].apply(em)
df.head()
```

```python
# How we can write the lambda Function!



df['Embarked'].apply(lambda x: 1 if x == 'S' else (2 if x == 'C' else 3))
```

```python
# Both are same way !
```

```python
# How to delete a column !

# We use the Function >>>> DROP

# > both rows and dolumns
# it takes row and column !

df.drop(index = 2) # drop the third Row
```

```python
# to make it permanent we use >> INPLACE >> THIS IS PERMANENT
df
```

```python
df.drop(index = 1 , inplace = True)
```

```python
df.head()
```

```python
# now we dont get the original data back
```

```python
# how to drop a columns
df.drop(columns = 'Double Price') # we can use the inplace here too !
```
