# **More Complex Functions**

**1. Grouping The Data**

```python
# Group the Data on gender To find the Average-Tenure

df.groupby('gender').agg({'tenure':'mean'})

# DATAFRAME > COLUMN WHICH NEED TO GROUP >> aggrigate on tenure column and do avg of it >>
```

```python
# GROUPING THE DATA >> multiple

df.groupby(['gender' ,'Churn']).agg({'tenure' : 'mean' , 'customerID' : 'count'})
```

```python
# PIVIT TABLE #

# >>   WE CAN GROUP DATA IN BOTH ROW AND COLUMN
# how to generate Pivit Table

df.pivot_table(index = 'gender' , columns = 'Churn' , values = 'tenure' , aggfunc = 'mean')
# Rows > Columns > Values > AggFunction
```

```python
# First we will create a function >> Conditions >>

# df['tenure'].apply(func-b)



# 1) Create a function that take input X
def t_b(x):
  if x < 5 :
    return "D"
  elif x>=5 and x<20:
    return "C"
  elif x>= 20 and x<50:
    return "B"
  else:
    return "A"


# 2) Applying this Function on the Tenure Column !

df['tenure'].apply(t_b)
```

```python
# 3) Assign this series to new Column !
df['newcol'] = df['tenure'].apply(t_b)
```

```python
# when we create a function it preserves in memory !  and we dont need this function for
# next usage ! onyl need one time !

# Better way >>>> LAMBDA FUNCTION !
import pandas as pd

# Convert 'tenure' column to numeric (if it's not already)
df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')

# Apply the lambda function to create a new column 'tenure_category'
df['tenure'].apply(
    lambda x: "D" if x < 5 else ("C" if x < 20 else ("B" if x < 50 else "A"))
)

```
